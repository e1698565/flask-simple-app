Simple Client

```python
import socket

client_sock = socket.create_connection(('127.0.0.1',8234))
client_sock.sendall("Hello".encode('utf-8'))
reply = client_sock.recv(4096)
print(f"Reply: {reply.decode()}")
client_sock.close()
```

Client Multiple Message

```python
import socket

client_sock = socket.create_connection(('127.0.0.1',8234))
while True:
  msg = input("Message to server: ")
  client_sock.sendall(msg.encode())
  reply = client_sock.recv(4096).decode()
  print(f"Reply: {reply}")
  if msg == "Bye":
    break
client_sock.close()
```

HTTP_Get

```python
import socket

client_sock = socket.create_connection(('127.0.0.1',8234))
req = '''GET /server.py HTTP/1.1

'''
client_sock.sendall(req.encode())
reply = client_sock.recv(4096).decode()
print(f"Reply: {reply}")
client_sock.close()
```
